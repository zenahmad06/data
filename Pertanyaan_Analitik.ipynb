# MENGGUNAKAN FRAMEWROK
kita bisa menggunakan framework dalam membangun analytical questions contohnya menggunakan SMART DA

misal sib questionnya di atas : apakah demografi mempengaruhi churn atau tidak ? kita buat analitycal questions dengan SMART-DA, contoh

terdiri dari

- Specific : fokus pada demografi contoh pada customer age atau gender, bisa aja menggunakan semua
- Measurement : menggunakan tingkat churn dalam persen per variabel
- Action-Oriented : temukan variabel dengan tingkat retensi dengan cara pengelompokan data segmen umur/gender
- Relevant : tujuan pertanyaan, untuk mengurangi churn
- time-bound : data waktu yang digunakan, pada dataset ini menggunakan 1 tahun data/12 bulan
